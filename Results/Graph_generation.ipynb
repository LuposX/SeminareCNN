# Importing needed Packages

```python
import json
import matplotlib.pyplot as plt
```

# Load the data

```python
file_name = 'val_loss_result_data'
```

```python
with open(file_name+ ".json") as user_file:
  file_contents = user_file.read()
```

```python
# Parse JSON data
parsed_data = json.loads(file_contents)
```

# Create the Plot

```python
# Iterate over the plots and create subplots
fig, ax = plt.subplots()

# Plot each scatter plot
for plot_data in parsed_data:
    x = plot_data['x']
    y = plot_data['y']
    name = plot_data['name']

    ax.yaxis.label.set_size(12)
    ax.xaxis.label.set_size(12)
    ax.plot(x, y, label=name)


# Add a legend
ax.legend(loc="upper left", bbox_to_anchor=(0.5, 1.1), fontsize=11)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

plt.grid(True, linestyle="--")

plt.xlabel("Validation Steps")
plt.ylabel("Validation Loss")

# Display the plot
plt.show()

ax.get_figure().savefig(file_name + ".png")
```
